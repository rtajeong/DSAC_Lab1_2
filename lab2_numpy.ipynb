```python
# 0~11 값을 얻고 이를 4x3 배열로 재구성한다
data = np.arange(12).reshape(4,3)
data
```

```python
data > 5
```

```python
# 5 보다 큰 값을 모두 5로 교체하는 방법
data[data > 5] = 5            # data 값이 바뀜에 주의
data
```

```python
# 행의 순서를 임의의 순으로 바꿀 수 있다
# 팬시 색인으로 배열을 바꾸면 복사본이 생기며 원래의 내용을 바뀌지 않는다
data[[3,2,0,1]]
```

```python
data
# data 내용은 그대로 남아 있다
```

```python
# 열의 순서도 바꿀 수 있다
data[:,[2,1,0]]
```

```python
# 행과 열을 모두 바꾸려면 아래와 같이 한다
print(data[[3,2,1,0]])
print(data[[3,2,1,0]][:,[2,1,0]])
```

```python
(data[[3,2,1,0]])[:,[2,1,0]]
```

```python
(data[[3,2,1,0]])[[2,1,0]]
```

```python
# 트랜스포즈는 행과 열의 위치를 바꾸는 것이다
data.T
```

```python
# 배열의 곱셈은 np.dot()을 사용한다
np.dot(data,data.T)
```

```python
data.dot(data.T)
```

```python
print(data.mean())
print(data.mean(axis=0)) # 각 열에 대한 평균
print(data.mean(0))      # axis 생략 가능
print(data.mean(1))      # 각 행에 대한 평균
```

```python
print(data.sum())
print(data.sum(0))     
print(data.sum(1))      
```

```python
# 배열에서 값이 특정 수보다 큰 원소의 갯수를 반환한다.
x = np.array([1,2,3,4,5,6])
print((x>4))
(x>4).sum()
```

```python
print(x[x>4])
x[x>4].sum()
```
