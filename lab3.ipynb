# Web scraping
- job 정보 가져오기

```python
url = 'https://www.monster.com/jobs/search/?q=Software-Engineer&where=MA'

page = requests.get(url)

# print(page.text)

soup = BeautifulSoup(page.text, 'html.parser')
# soup

result = soup.find(id="SearchResults")

job_elems = result.find_all('section', class_='card-content')
```

```python
len(job_elems)
```

```python
job_elems[0]
```

```python
job_elems[0].find('h2', class_='title').text.strip()
```

```python
job_elems[0].find('div', class_='company').text.strip()
```

```python
job_elems[0].find('div', class_='location').text.strip()
```

```python
# 모든 것에 대해서
for job_elem in job_elems:
    title = job_elem.find('h2', class_='title')
    company = job_elem.find('div', class_='company')
    location = job_elem.find('div', class_='location')
    if None in (title, company, location):
        continue
    print(title.text.strip())
    print(company.text.strip())
    print(location.text.strip())
    print()
```

```python
# complete code
# url = 'https://www.monster.com/jobs/search/?q=Software-Engineer&where=MA'
url = 'https://www.monster.com/jobs/search/?q=Software-Engineer&where=MA&stpage=1&page=10'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
result = soup.find(id="SearchResults")
job_elems = result.find_all('section', class_='card-content')
print(len(job_elems))
for job_elem in job_elems:
    title = job_elem.find('h2', class_='title')
    company = job_elem.find('div', class_='company')
    location = job_elem.find('div', class_='location')
    if None in (title, company, location):
        continue
    print(title.text.strip())
    print(company.text.strip())
    print(location.text.strip())
    print()
```
